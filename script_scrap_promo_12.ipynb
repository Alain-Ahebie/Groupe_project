```python
class Script_promo_12:
    
    def __init__(self,betclic,pmu,france_pari,parion_sport,netbet,zebet,betstars,william):
        
        self.betclic = betclic
        self.pmu = pmu
        self.france_pari = france_pari
        self.parion_sport = parion_sport
        self.netbet = netbet
        self.zebet = zebet
        self.betstars = betstars
        self.william = william

###  script Nicolas ###

    def nicolas(self):
        
        url = self.betclic
        r = requests.get(url)
        soup = BeautifulSoup(r.content,'html')
        result_f = []
        result_final = []
        tmp = []
        model = '\w+\_\d{7}'
        for elem in soup.find_all('div'):
            tmp.append(elem.get('id'))
        tmp1=str(tmp)
        result_f.append(re.findall(model,tmp1))
        for e in result_f[0]:
            if 'match' in e:
                result_final.append(e)
            else:
                continue
        test=[]
        for ele in result_final:
            for elem in soup.find_all('div',attrs={'id':ele}):
                for el in elem.text.strip().split('\n'):
                    test.append(el.replace('                \r','').replace('\r','').replace('            ',''))
        
        result=[]
        for elem in test:
            if len(elem) > 2:
                result.append(elem)
                
        liste = []
        for elem in result:
            if '+' in elem or 'paris' in elem or '    ' in elem:
                continue
            else:
                liste.append(elem)
                
        match = []
        un = []
        nul = []
        deux = []
        for i in range(len(liste)):
            if i % 4 == 0:
                match.append(liste[i])
            if i % 4 == 1:
                un.append(liste[i])
            if i % 4 == 2:
                nul.append(liste[i])
            if i % 4 == 3:
                deux.append(liste[i])
                
        df = pd.DataFrame({'match':(match),'1':(un),'nul':(nul),'2':(deux)})
        un = []
        for elem in df['1']:
            un.append(elem)
        nul = []
        for elem in df['nul']:
            nul.append(elem)
        deux = []
        for elem in df['2']:
            deux.append(elem)
        domicile = []
        visiteur = []
        for elem in match:
            domicile.append(elem.split('-')[0])
            visiteur.append(elem.split('-')[1])
            
        liste_tmp = []
        for i in range(len(domicile)):
            dico = {}
            dico[domicile[i]]=un[i]
            dico['nul']=nul[i]
            dico[visiteur[i]]=deux[i]
            liste_tmp.append(dico)
        dico_nicolas = {}
        dico_nicolas['ligue_1_nicolas']=liste_tmp
                
        return dico_nicolas
    
    # script de richard

    def richard(self):

        url=self.france_pari
        scr = requests.get(url)
        soup = BeautifulSoup(scr.content,'html')

        cote_1 = []
        cote_nul = []
        cote_2 = []

        for elem in soup.find_all('div',attrs={"labelnum":"1"}):
            cote_1.append(elem.text)

        for elem in soup.find_all('div',attrs={"labelnum":"N"}):
            cote_nul.append(elem.text.replace("N", ""))

        for elem in soup.find_all('div',attrs={"labelnum":"2"}):
            cote_2.append(elem.text)

        equipes_A = []
        equipes_B = []

        for elem in soup.find_all('span',attrs={"class" :"bet-libEvent"}):
            equipes_A.append(elem.text.strip().split("/")[0])
            equipes_B.append(elem.text.strip().split("/")[1])

        liste = []
        for i in range(len(equipes_A)):
            liste.append({equipes_A[i]:cote_1[i] , "Null":cote_nul[i], equipes_B[i]:cote_2[i]})
        dico_richard = {}
        dico_richard['ligue_1_richard']=liste

        return dico_richard

    ### script alain ###
    
    def alain(self):
        
        headers = {'Accept' :'application/json, text/plain, */*'.encode('cp1252'),
        'Accept-Encoding':'gzip, deflate, br'.encode('cp1252'),
        'Accept-Language':'fr-FR'.encode('cp1252'),
        'Cache-Control':'max-age=0'.encode('cp1252'),
        'Connection':'keep-alive'.encode('cp1252'),
        'Cookie':'abp-pselw=abpwfr11006_1443; device_view=full; fdj_status_cookie=true; TCPID=11931104559846318062; atuserid=%7B%22name%22%3A%22atuserid%22%2C%22val%22%3A%22fc25f63d-1361-4385-8a83-648840dc31e6%22%2C%22options%22%3A%7B%22end%22%3A%222020-04-18T09%3A45%3A59.438Z%22%2C%22path%22%3A%22%2F%22%7D%7D; atidvisitor=%7B%22name%22%3A%22atidvisitor%22%2C%22val%22%3A%7B%22vrn%22%3A%22-484303-%22%2C%22at%22%3A%22%22%2C%22ac%22%3A0%7D%2C%22options%22%3A%7B%22path%22%3A%22%2F%22%2C%22session%22%3A15724800%2C%22end%22%3A15724800%7D%7D; cikneeto_uuid=id:cf2bc69e-3470-4040-aa5d-1e8e0a45c874; cikneeto=date:1552905064863; fdj-pac=%7B%22appName%22%3A%22PORTAIL_WEB_PW%22%2C%22appVersion%22%3A%221516%22%2C%22pushInfo%22%3A%22%22%7D; fdj_privacy_cookie=true; TC_OPTOUT=0@@@@@@ALL'.encode('cp1252'),
        'Host':'www.enligne.parionssport.fdj.fr'.encode('cp1252'),
        'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:65.0) Gecko/20100101 Firefox/65.0'.encode('cp1252'),
        'X-LVS-HSToken':'ktNVYXgYJAEgn-80jGBesRtSL98_aOGqNsFJgTKqZPFZGney6O8kSIpkfeNCCgcbX8DGhCiaUVSq_cS88gbRwvRs1ylbD_Rc2Ysqey_O0Y2GaB38hdM9vQzEQvibcNGwOcF-6u_qQE6WlhKNLeBN8w=='.encode('cp1252'),
        }
        url=self.parion_sport
        r = requests.get(url , headers = headers )
        json.loads(r.text)
        x=json.loads(r.text)
        xjson = x['items']
        Parent_n = []
        for elem in xjson :
            if "m" in elem :
                Parent_n.append(elem)
        keys_e = []
        for elem in xjson :
            if "e" in elem :
                keys_e.append(elem)
        dico_e = {}
        for elem in xjson :
            for k  in keys_e :
                if elem == k :
                    dico_e[k] = xjson.get(elem).get('desc')
        dico_e = {}
        for elem in xjson :
            for k in Parent_n :
                if k in  xjson.get(elem).get('parent') :
                    dico_e[k] =  xjson.get(elem).get('desc')
        match = []
        Team = {}
        Nul = {}
        for elem in xjson :
            if 'p' in  xjson.get(elem).get('parent'):
                match.append(xjson.get(elem).get('desc'))
            if 'm' in  xjson.get(elem).get('parent'):
                Team[xjson.get(elem).get('desc')] = xjson.get(elem).get('price')
                if "N" in xjson.get(elem).get('desc'):
                    Nul[dico_e.get(xjson.get(elem).get('parent'))] = xjson.get(elem).get('price')
        match.remove('Ligue 1 Conforama')
        match
        n = []
        for elem in match:
            n.append(elem.split("vs"))
        l = []
        for elem in n :
            dic = {}
            for el in elem :
                dic[el] = Team[el.strip()]
                if el.strip() in Nul :
                    dic["Nul"] = Nul[el.strip()]
            l.append(dic)
        dico_alain = {}
        dico_alain["ligue_1_alain"] = l
        
        return dico_alain
    
    ### script Tashi ###
    
    def parse_page(url):
        r = requests.get(url)
        html = r.text.strip()
        soup = BeautifulSoup(html, 'lxml')


        away, Away_win = Script_promo_12.extract_away_win(soup)

        Nul = Script_promo_12.draws(soup)

        home, Home_win = Script_promo_12.extract_home_win(soup)


        detail = []

        for i in range(len(Home_win)):
            detail.append({home[i]:Home_win[i],"Nul":Nul[i],away[i]:Away_win[i]})

        #saving as a pickle file 

        pickle.dump(detail,open('Cote_France-Paris','wb'))
    
        return detail
 
        # extracting home win cotes

    def extract_home_win(soup):
        home = []
        Home_win =[]
        for i in soup.find_all("div", attrs={"class":"obet-event-list-grid-formatter-col col-rows col-xs-3"}):
            for e in i.find_all("em", attrs={"class":"trow--event--name"}):
                home.append(e.text.strip().split("//")[0])
        for e in soup.find_all("li", attrs={"class":"col-sm-3 trow--odd--item"}):
            for e in e.find_all('a', attrs={"data-minor-sort":"H"}):
                Home_win.append(e.text.strip())



        return home, Home_win

    # extracting away win cotes

    def extract_away_win(soup):
        away = []
        Away_win = []

        for i in soup.find_all("div", attrs={"class":"obet-event-list-grid-formatter-col col-rows col-xs-3"}):
            for e in i.find_all("em", attrs={"class":"trow--event--name"}):
                away.append(e.text.strip().split("//")[1])
        for e in soup.find_all("li", attrs={"class":"col-sm-3 trow--odd--item"}):
            for e in e.find_all('a', attrs={"data-minor-sort":"A"}):
                Away_win.append(e.text.strip())

        return away, Away_win


    # extracting all the draw cotes
    def draws(soup):
        Nul = []
        for e in soup.find_all("li", attrs={"class":"col-sm-3 trow--odd--item"}):
            for e in e.find_all('a', attrs={"data-minor-sort":"D"}):
                Nul.append(e.text.strip())

        return Nul
    
    def tashi(self):
        
        url = self.pmu
        wmn_exp = Script_promo_12.parse_page(url)
        dico_tashi = {}
        dico_tashi['ligue_1_tashi']=wmn_exp
        
        return dico_tashi
    
    ### script Marianna ###
    
    def marianna(self):
        
        url=self.netbet
        r = requests.get(url)
        soup = BeautifulSoup(r.content,'html')
        
        equipes1 = []
        equipes2 = []
        for elem in soup.find_all('a', attrs={"class" :"uk-margin-right uk-text-truncate"}):
            try:
                equipes1.append(elem.text.strip().split(' / ')[0])
            except:
                equipes1.append(None)
            try:
                equipes2.append(elem.text.strip().split(' / ')[1])
            except:
                equipes2.append(None)
                
        domicile = []
        exterieur = []
        Null = []
        liste = soup.find_all('a',attrs={"class" :"nb-load uk-flex"})
        parie =[]
        n=0

        for i in range(0,len(liste),6):
             parie.append(liste[i:i+6][:3])
        for elem in parie:

            domicile.append(elem[0].text)
            Null.append(elem[1].text)
            exterieur.append(elem[2].text)
            if n == len(equipes1)-1:
                break
            n+=1
        ligue_1 = []
        for i in range(len(equipes1)):
            ligue_1.append({equipes1[i]:domicile[i], "Null":Null[i], equipes2[i]:exterieur[i]})
            
        dico_marianna = {}
        dico_marianna['ligue_1_marianna']=ligue_1
        
        return dico_marianna
    
    ### script Ary ###
    
    def ary(self):
        
        url= self.zebet
        r = requests.get(url)
        bet = BeautifulSoup(r.content,'html')
        
        cote = []
        for a in bet.find_all("span", attrs = {"class":"pmq-cote uk-width-1-1"}):
            cote.append(a.text)
        
        t1 = []
        t2 = []
        for t in bet.find_all("div", attrs ={"class":"uk-visible-small uk-text-bold uk-margin-left uk-text-truncate"}):
            t1.append(t.text.split('/')[0].strip())
        for e in bet.find_all("div", attrs ={"class":"uk-visible-small uk-text-bold uk-margin-left uk-text-truncate"}):
            t2.append(e.text.split('/')[1].strip())
            
        date = []
        heure = []
        for d in bet.find_all("div", attrs={"class":"bet-time"}):
            date.append(d.text.split(" ")[0])
        for h in bet.find_all("div", attrs={"class":"bet-time"}):
            heure.append(h.text.split(" ")[1])
            
        team1 = cote[0::3]
        nul = cote[1::3]
        team2 = cote[2::3]
    
        dico_ary = {}
        dico_ary["Ligue_1_ary"] = []
        for i in range(len(t1)):
            dico_ary.get("Ligue_1_ary").append({t1[i]:team1[i], "Nul":nul[i], t2[i]:team2[i], "Date":date[i], "Heure":heure[i]})

        return dico_ary
    
    def lock(self):
        
        url = self.betstars
        r = requests.get(url)
        a = json.loads(r.text)
        i=0
        matchs=[]
        for elem in a['event']:
            matchs.append(a['event'][i].get('name'))
            i = i+1
        i=0
        equipe_domicile=[]
        equipe_externe=[]

        for elem in matchs:
            equipe_domicile.append(matchs[i].split(' vs ')[0])
            equipe_externe.append(matchs[i].split(' vs ')[1])
            i = i+1
        i=0
        cote1=[]

        for elem in a['event']:
            cote1.append(a['event'][i].get('markets')[0].get('selection')[0].get('odds'))
            i = i+1
        i=0
        cote_exterieur=[]

        for elem in cote1:
            cote_exterieur.append(cote1[i].get('dec'))
            i = i+1
        i=0
        cote2=[]

        for elem in a['event']:
            cote2.append(a['event'][i].get('markets')[0].get('selection')[1].get('odds'))
            i = i+1
        i=0
        cote_nul=[]

        for elem in cote2:
            cote_nul.append(cote2[i].get('dec'))
            i = i+1     
        i=0
        cote3=[]

        for elem in a['event']:
            cote3.append(a['event'][i].get('markets')[0].get('selection')[2].get('odds'))
            i = i+1
        i=0
        cote_domicile=[]

        for elem in cote3:
            cote_domicile.append(cote3[i].get('dec'))
            i = i+1
        i=0
        cote_domicile=[]

        for elem in cote3:
            cote_domicile.append(cote3[i].get('dec'))
            i = i+1
        dico={}
        Ligue1=[]
        i = 0

        for i in range(len(equipe_domicile)):
            dico[equipe_domicile[i]] = cote_domicile[i]
            dico["Nul"] = cote_nul[i]
            dico[equipe_externe[i]] = cote_exterieur[i]
            Ligue1.append(dico)
            dico={}
        dico_lock = {}
        dico_lock['ligue_1_lock']=Ligue1
        
        return dico_lock
    
    def oumar(self):
        
        url= self.william
        r = requests.get(url)
        ligue1 = BeautifulSoup(r.content,'html')
        for elem in ligue1.find_all('tbody',attrs={'id':'tup_mkt_grp_tbl_UC_9d8a08d4b13c912153e27659829a27ad'}):
            a=" ".join(elem.text.split()).split('+') 
        date=[]
        infos_equipe=[]
        for elem in a:
            try:
                infos_equipe.append(elem.split('UK')[1].strip())

            except:
                pass
            try:
                date.append(elem.split('UK')[0].strip())
            except:
                pass

        new_date = []
        for ele in date:
            if len(ele) > 12:
                new_date.append(ele[len(ele)-12:])
            elif len(ele) < 12:
                new_date.append("No infos")
            else:
                new_date.append(ele)
        dom = []
        ext = []
        Null = []
        dom_score = []
        ext_score = []
        for elem in infos_equipe:
            for i in range(len(elem)):
                if elem[i].isnumeric():
                    tmp1 = elem[:i].split("₋")[0].strip() #Nom de l'équipe domicile

                    tmp2 = elem[:i].split("₋")[1].strip()#Nom de l'équipe extérieur

                    dom1 = elem[i:].split()[0].strip()
                    null = elem[i:].split()[1].strip()
                    ext1 = elem[i:].split()[2].strip()
                    dom.append(tmp1)
                    ext.append(tmp2)
                    dom_score.append(dom1)
                    ext_score.append(ext1)

                    Null.append(null)
                    break   
        dico_oumar = {}
        ligue1 = []
        for i in range(len(infos_equipe)):
            ligue1.append({"date":new_date[i],dom[i]:dom_score[i],"Null":Null[i],ext[i]:ext_score[i]})
        dico_oumar["ligue_1_oumar"] = ligue1 
        
        return dico_oumar
    
        
class Recup_dico:
    
    def __init__(self,d1,d2,d3,d4,d5,d6,d7,d8):
        
        self.dico_1 = d1
        self.dico_2 = d2
        self.dico_3 = d3
        self.dico_4 = d4
        self.dico_5 = d5
        self.dico_6 = d6
        self.dico_7 = d7
        self.dico_8 = d8
        
    def append_dico(self):
        
        liste_dico = []
        
        liste_dico.append(self.dico_1)
        liste_dico.append(self.dico_2)
        liste_dico.append(self.dico_3)
        liste_dico.append(self.dico_4)
        liste_dico.append(self.dico_5)
        liste_dico.append(self.dico_6)
        liste_dico.append(self.dico_7)
        liste_dico.append(self.dico_8)
        
        return liste_dico
    
    
```
